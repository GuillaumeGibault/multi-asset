```python
def transform_monthly(df):
    df = df.reset_index(drop = True) # Just in case index is not 0 to len(df)-1
    for index, row in df.iterrows():
        if index == 0:
            df_m = df.iloc[[index]] # Create a new dataframe
    
        if index != len(df)-1: # If it is not the last row as index + 1 would be impossible
            if str(row['Date'])[0:7] != str(df['Date'][index + 1])[0:7]: # If yyyy-mm of row date != yyyy-mm of next row --> either month or year !=
                df_m = df_m.append(df.iloc[[index]]) #Add last day of the month to the new dataframe
        
    df_m = df_m.drop(index = 0).reset_index(drop = True)
    return df_m
```

```python
def mean_vol(df, period):
    if period == 'daily':
        df_mean = pd.DataFrame(df.mean()*250*100).T
        df_std = pd.DataFrame(df.std()*(250**0.5)*100).T
    
    if period == 'monthly':
        df_mean = pd.DataFrame(df.mean()*12*100).T
        df_std = pd.DataFrame(df.std()*(12**0.5)*100).T        
    
    df_mean = df_mean.rename(index={0: 'mean'})
    df_std = df_std.rename(index={0: 'std'})
    df_final = df_mean.append(df_std)
    return df_final
```

```python
def mean_ann(df, period): # Period must be either 'daily' or 'monthly'
    if period == 'daily':
        df_mean = df.mean()*250
    if period == 'monthly':
        df_mean = df.mean()*12
    return df_mean
```

```python
def var_cov_ann(df, period):
    if period == 'daily':
        df_cov = df.cov()*250
    if period == 'monthly':
        df_cov = df.cov()*12
    return df_cov
```

```python
def ptf_metrics(df_returns, period, weights): # Period either 'daily' or 'monthly', weights as np.array
    # Make the sum of weights = 1
    weights = weights / np.sum(weights)

    # PTF expected returns
    exp_ret = np.sum(mean_ann(df_returns, period) * weights)

    # PTF expected volatility
    exp_vol = np.sqrt(np.dot(weights.T, np.dot(var_cov_ann(df_returns, period), weights)))

    # Compute Sharpe ratio
    sharpe = (exp_ret - mean_ann(df_returns, period)['Rf']) / exp_vol

    weights = np.round(weights * 100, 2)
    exp_ret = np.round(exp_ret * 100, 2)
    exp_vol = np.round(exp_vol * 100, 2)

    df_results = pd.DataFrame([sharpe, exp_ret, exp_vol, weights]).T
    df_results.columns = ['Sharpe', 'Expected return', 'Expected vol', 'Weights']

    return df_results
```

# Portfolio Optimization

```python
def ptf_opti(ptf_nb, df_returns, period, min_ret=0, max_vol=20): # min_ret & max_vol have default values
                                                                  # can change them when calling the function
    assets_nb = len(df_returns.columns.to_list())
    rf = mean_ann(df_returns, period)['Rf']
     
    weights_arr = np.zeros((ptf_nb, assets_nb)) # Weights array
    ret_arr = np.zeros(ptf_nb) # Returns array
    vol_arr = np.zeros(ptf_nb) # Vol array
    sharpe_arr = np.zeros(ptf_nb) # Sharpe array
    
    for i in range(ptf_nb):
        # Generate random weights
        weights = np.array(np.random.random(assets_nb))
        
        # Make the sum of weights = 1
        weights = weights / np.sum(weights)
        
        # Add the weights to the weights array
        weights_arr[i, :] = weights
        
        # PTF expected returns
        ret_arr[i] = np.sum(mean_ann(df_returns, period) * weights)
        
        # PTF expected volatility
        vol_arr[i] = np.sqrt(np.dot(weights.T, np.dot(var_cov_ann(df_returns, period), weights)))

        # Compute Sharpe ratio
        sharpe_arr[i] = (ret_arr[i] - rf) / vol_arr[i]
        
        # Clean results
        weights_arr[i] = np.round(weights_arr[i, :] * 100, 2)
        ret_arr[i] = np.round(ret_arr[i] * 100, 2)
        vol_arr[i] = np.round(vol_arr[i] * 100, 2)
        sharpe_arr[i] = np.round(sharpe_arr[i], 2)
    
    # Combine metrics in a DataFrame
    df_res = pd.DataFrame([sharpe_arr, ret_arr, vol_arr, weights_arr]).T
    
    # Take out best ptf respecting conditions
    df_res.columns = ['Sharpe', 'Expected return', 'Expected vol', 'Weights']
    df_res = df_res[df_res['Expected return'] > min_ret]
    df_res = df_res[df_res['Expected vol'] < max_vol]
    df_res = df_res.sort_values(by = ['Sharpe'], ascending = False)
    
    pd.set_option('display.max_colwidth', None)
    
    return df_res.head(10)
```
